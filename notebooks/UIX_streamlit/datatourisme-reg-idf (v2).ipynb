#### 4.1. Nettoyer la colonne Categories_de_POI → passer d’URI à labels lisibles + liste de catégories

```python
#décrire la colonnne Categories_de_POI
print("\n-------------------------------------")
print("Les valeurs uniques de la colonne Categories_de_POI :")
print("---------------------------------------")
display(df_gov["Categories_de_POI"].unique()[:5])
```

```python
#créer la colonne Categories_de_POI
#fonction : nettoyer → passer d’URI à labels lisibles + liste de catégories
def uri_to_label(uri: str) -> str:
    """Extrait le label final d'une URI (après # ou /)."""
    if pd.isna(uri) or uri is None:
        return np.nan
    uri = str(uri).strip()
    # coupe sur # sinon sur /
    if "#" in uri:
        return uri.split("#")[-1]
    return uri.rstrip("/").split("/")[-1]

def split_categories(cell: str):
    """Split sur | puis transforme chaque URI en label."""
    if pd.isna(cell) or cell is None:
        return []
    parts = [p.strip() for p in str(cell).split("|") if p.strip()]
    labels = [uri_to_label(p) for p in parts]
    # dédoublonnage en gardant l'ordre
    seen = set()
    out = []
    for x in labels:
        if x not in seen and x is not np.nan:
            seen.add(x)
            out.append(x)
    return out

df_gov["categories_list"] = df_gov["Categories_de_POI"].apply(split_categories)

#vérifier
df_gov[["Categories_de_POI", "categories_list"]].head()
```

#### 4.2. Nettoyer la colonne Classements_du_POI → Extraire le nombre d’étoiles

```python
print("\n-------------------------------------")
print("Les valeurs uniques de la colonne Classements_du_POI :")
print("---------------------------------------")
display(df_gov["Classements_du_POI"].unique()[:10])
```

```python
#Extraire le nombre d’étoiles (regex robuste)
def extract_stars(value):
    if pd.isna(value):
        return None

    text = str(value)

    # cherche "X étoile" ou "X étoiles"
    match = re.search(r'([1-5])\s*étoiles?', text)
    if match:
        return int(match.group(1))
    return None

#vérifier la colonne rating_stars
df_gov["rating_stars"] = df_gov["Classements_du_POI"].apply(extract_stars)
rating_stars = df_gov["rating_stars"].value_counts()
display(rating_stars)

rating_stars_isna = df_gov["rating_stars"].isna().value_counts()
rating_stars_isna
```

#### 4.3. Séparer la colonne Code_postal_et_commune en 2 colonnes : Code_postal, commune

```python
# Séparation Code_postal_et_commune
col = "Code_postal_et_commune"
parts = (df_gov[col].astype(str).str.split("#", n=1, expand=True))

df_gov["postal_code"] = parts[0].str.strip()
df_gov["commune"] = parts[1].str.strip()

# Vérification
df_gov[["postal_code", "commune"]].head()
```

#### 4.6. Proposer des zones/quartiers recommandés

```python
#Crée colonne zone_id (grille géographique)
def compute_zone_id(lat, lon, cell_size_km=1.0):
    # approx : 1° latitude ≈ 111 km
    cell_deg = cell_size_km / 111.0
    lat_id = np.floor(lat / cell_deg).astype(int)
    lon_id = np.floor(lon / cell_deg).astype(int)
    return lat_id.astype(str) + "_" + lon_id.astype(str)

df_gov = df_gov.copy()
df_gov["zone_id"] = compute_zone_id(df_gov["Latitude"], df_gov["Longitude"], cell_size_km=1.0)
df_gov[["Latitude","Longitude","zone_id"]].head()
```

```python
# scoring les zones/quartiers recommandés¶
zones = (df_gov.groupby(["zone_id", "type"]).size().unstack(fill_value=0))
zones["housing_score"] = zones.get("Hotel", 0)
zones["restaurant_score"] = zones.get("Food", 0)
zones["shopping_score"] = zones.get("Shopping", 0)
zones["culture_score"] = zones.get("Museum", 0) + zones.get("Monument", 0)
zones["life_score"] = zones.get("Evenement", 0) + zones.get("Cinema", 0) + zones.get("Healthcare", 0)

#Calculer le centroïde de chaque zone
zone_centroids = (df_gov.groupby("zone_id").agg(lat=("Latitude", "mean"), lon=("Longitude", "mean"),))

#Fusionner scores + coordonnées
zones_reco = zones.merge(zone_centroids, left_index=True, right_index=True)
```

```python
#Normaliser chaque score
def normalize(series):
    if series.max() == series.min():
        return series * 0
    return (series - series.min()) / (series.max() - series.min())

for col in ["housing_score", "restaurant_score", "shopping_score", "culture_score", "life_score"]:
    zones[f"{col}_n"] = normalize(zones[col])


#Calcul du final_score_quart_reco --> poids housing=2.0, culture=1.2, restaurant=1.0, shopping=0.7, life=0.6
zones["final_score_quart_reco"] = (
    1.2 * zones["housing_score_n"] +
    1.0 * zones["culture_score_n"] +
    0.8 * zones["life_score_n"] +
    0.6 * zones["restaurant_score_n"]
)
zones.sort_values("final_score_quart_reco", ascending=False).head(3)
```

```python
#Sélectionner les “meilleures zones” : Objectif : ne garder que les zones pertinentes pour l’utilisateur.
TOP_N = 10
top_housing = zones.sort_values("housing_score_n", ascending=False).head(10)
top_culture = zones.sort_values("culture_score_n", ascending=False).head(10)
top_life = zones.sort_values("life_score_n", ascending=False).head(10)
top_food = zones.sort_values("restaurant_score_n", ascending=False).head(10)
top_zones = (pd.concat([top_housing, top_culture, top_life, top_food]).drop_duplicates().reset_index())


#Ajouter les coordonnées (indispensable pour la carte)
zone_centroids = (df_gov.groupby("zone_id").agg(lat=("Latitude", "mean"),lon=("Longitude", "mean"),))
top_zones = top_zones.merge(zone_centroids,on="zone_id",how="left")

#Donner un sens UX aux zones
def label_zone(row):
    if row["housing_score"] >= row["culture_score"]:
        return "Idéal pour se loger"
    elif row["culture_score"] >= row["life_score"]:
        return "Quartier culturel"
    else:
        return "Quartier vivant"

top_zones["label"] = top_zones.apply(label_zone, axis=1)

#vérifier :
display(top_zones[["zone_id", "final_score_quart_reco", "label"]])
```

```python
#Afficher ces zones sur une carte (Jupyter)
import folium

m = folium.Map(location=[top_zones.lat.mean(), top_zones.lon.mean()],zoom_start=13)

for _, r in top_zones.iterrows():
    folium.Marker(
        [r.lat, r.lon],
        popup=f"""
        <b>{r.label}</b><br>
        Score : {r.final_score_quart_reco:.2f}
        """,
        icon=folium.Icon(color="green", icon="home")
    ).add_to(m)
m
```
