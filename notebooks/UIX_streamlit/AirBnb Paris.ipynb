```python
#transformer en booléen les colonnes require_guest_phone_verification (0% missing), require_guest_profile_picture (0% missing), 
''''
BOOL_COLS = ["require_guest_phone_verification", "require_guest_profile_picture", "host_is_superhost", "host_identity_verified"]
for col in BOOL_COLS:
    if col in df_listings.columns:
        df_listings[col] = (df_listings[col].astype(str).str.lower().isin(["t", "true", "1", "yes"]))
''''
```

```python
#transformer en Dates : last_review (21% missing), first_review (21% missing), host_since (0,01% missing)
''''
DATE_COLS = ["last_review", "first_review", "host_since"]
for col in DATE_COLS:
    if col in df_listings.columns:
        df_listings[col] = pd.to_datetime(df_listings[col], errors="coerce")
''''
```

```python
#transformer en colonnes catégorielles (optimisation mémoire) : room_type (% missing), property_type (% missing),
# bed_type (0% missing), neighbourhood_cleansed (% missing), cancellation_policy (0% missing)
''''
CAT_COLS = ["room_type", "property_type", "bed_type", "cancellation_policy"]
for col in CAT_COLS:
    if col in df_listings.columns:
        df_listings[col] = df_listings[col].astype("category")
''''
```

```python
#transformer en liste : amenities (0% missing)
'''
df_listings["amenities_list"] = (df_listings["amenities"].astype(str).str.strip("{}").str.replace('"', "").str.split(","))
'''
```

```python
#transformer en capacités types numériques propres : accommodates (0% missing), bedrooms (0.11% missing), beds (% missing),
#bathrooms (0.13% missing), minimum_nights (0% missing), maximum_nights (0% missing), guests_included (0% missing)

INT_COLS = ["accommodates", "bedrooms", "beds", "bathrooms", "minimum_nights", "maximum_nights", "guests_included"]
for col in INT_COLS:
    if col in df_listings.columns:
        df_listings[col] = pd.to_numeric(df_listings[col], errors="coerce")
```

```python
#convertir review_scores_rating sur 5
df_listings["review_scores_rating"] = (pd.to_numeric(df_listings["review_scores_rating"], errors="coerce") / 20).round(2)
```

```python
#Nettoyer + transformer en float : price (0% missing), cleaning_fee (27% missing), security_deposit (31% missing),
#extra_people (0% missing), weekly_price (79% missing), monthly_price (87% missing)

PRICE_COLS = ["price", "cleaning_fee", "security_deposit", "extra_people", "weekly_price", "monthly_price"]
for col in PRICE_COLS:
    if col in df_listings.columns:
        df_listings[col] = (df_listings[col].astype(str).str.replace(r"[^0-9.]", "", regex=True))
        df_listings[col] = pd.to_numeric(df_listings[col], errors="coerce")
        df_listings[col] = df_listings[col].astype("float").fillna(0)
```

```python
#créer et calculer la colonne "price_level" en mode quantiles basant sur colonne "price" : 1=petit <= q1 ; q1 < 2=moyen <= q2 ; 3=confort > q2

#regarder la distribution
df_listings["price"].describe(percentiles=[0.33, 0.66])

#calculer les seuils
q1 = df_listings["price"].quantile(0.33)
q2 = df_listings["price"].quantile(0.66)
q1, q2

#créer price_level
df_listings["price_level"] = np.nan

df_listings.loc[df_listings["price"] <= q1, "price_level"] = 1
df_listings.loc[(df_listings["price"] > q1) & (df_listings["price"] <= q2), "price_level"] = 2
df_listings.loc[df_listings["price"] > q2, "price_level"] = 3

df_listings[["price_level", "price"]].head(5)
```
